```python
class Person():
    def __init__(self, name, age):
        self.name=name
        self.age=age
        
    def details(self):
        print(f" My name is: {self.name} , and I am: {self.age} years old")
        
Person_1= Person("Alice", 25)

Person_1.details()
    
```
